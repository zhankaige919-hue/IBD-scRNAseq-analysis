```python
import scanpy as sc

adata = sc.read_h5ad("GSE309239_P1_SILP_processed.h5ad")

adata
```

```python
# PCA
sc.tl.pca(adata)

# neighbors
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=30)

# t-SNE
sc.tl.tsne(adata)

# plot
sc.pl.tsne(
    adata,
    color="leiden",
    legend_loc="right margin"
)
```
